```python
from transformers import pipeline
sentiment_analysis_pipeline = pipeline('sentiment-analysis', model="cardiffnlp/twitter-roberta-base-sentiment-latest")
result = sentiment_analysis_pipeline("World is becoming beatiful")

print(result)
print(f"sentimenet - {result[0]['label']} | Score - {result[0]['score']}")
```

```python
classifier = pipeline("zero-shot-classification",
                      model="facebook/bart-large-mnli")

sequence_to_classify = "Anura Kumara Dissanayake is The 9th precident of sri lanka"
candidate_labels = ['travel', 'cooking', 'dancing', 'politics']
classifier(sequence_to_classify, candidate_labels)
```

ner_pipeline = pipeline("ner" , model="dslim/bert-base-NER" , grouped_entities=True)

```python
qa_pipeline = pipeline("question-answering", model="deepset/roberta-base-squad2")
context = "Anura Kumara Dissanayake is the president of Sri Lanka"
question = "Who is the president of sri lanka"

result = qa_pipeline(question=question, context=context)
print(result)
```

```python
text_generator = pipeline("text-generation", model="openai-community/gpt2")
prompt = "Best prices to visit in Sri Lanka"
result = text_generator(prompt, max_length=100, temperature=0.7)
print(result)
```

```python
object_detection = pipeline("object-detection", model="facebook/detr-resnet-50")
image_url = "https://imgcdn.zigwheels.lk/large/gallery/exterior/10/73/land-rover-range-rover-front-angle-low-view-824577.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)
results = object_detection(image)
for result in results:
  print(f"Object - {result['label']} | Score - {result['score']}")
```

```python
from langchain_core.prompts import ChatPromptTemplate
from langchain_openai import ChatOpenAI
from google.colab import userdata
from pprint import pprint
key = userdata.get('OPENAI_API_KEY').strip()
model = ChatOpenAI(model="gpt-5.2-2025-12-11", api_key=key)

image_captioning = pipeline("image-to-text", model="Salesforce/blip-image-captioning-base")
image_url = "https://imgcdn.zigwheels.lk/large/gallery/exterior/10/73/land-rover-range-rover-front-angle-low-view-824577.jpg"
image = Image.open(requests.get(image_url, stream=True).raw)
text_result = image_captioning(image)

enhancer_prompt = ChatPromptTemplate.from_messages([
"""You are a social media marketing expert. Your task is to transform the provided text into an engaging, concise, and attention-grabbing social media caption that is optimized for reach and engagement.

Input text:
{text_result}"""
])
enhancer_chain = enhancer_prompt | model

result = enhancer_chain.invoke({"text_result" : text_result})
pprint(result.content)
```
